# étude des features quantitatives

sns.distplot()


```python
#features quantitatives : cherchons à mieux les comprendre (elles ne sont pas normalisées et suivent souvent une distribution gaussienne)
for col in df.select_dtypes('float'):
    plt.figure()
    sns.distplot(df[col])
```

# Analyse des relations target/features

# on crée deux sous-groupes selon le résultat au test final
booléens sur les df

```python
df_positive=df[df['Binar OP130_Resultat_Global_v']==1]
df_negative=df[df['Binar OP130_Resultat_Global_v']==0]
```

# lien target/features

```python
#FEATURES QUANTITATIVES : peu de différences notables!


for col in df.select_dtypes('float'):
    plt.figure()
    sns.distplot(df_positive[col],label='positive')
    sns.distplot(df_negative[col],label='negative')
    plt.legend()
```
