# (Quantitative) features distribution.

```python
#features : mostly seem to follow a gaussian distribution
for col in df.select_dtypes('float'):
    plt.figure()
    sns.distplot(df[col])
```

# Target/features relationship :

# Creating one group for each class
(boolean indexing on df)

```python
df_positive=df[df['Binar OP130_Resultat_Global_v']==1]
df_negative=df[df['Binar OP130_Resultat_Global_v']==0]
```

```python
#FEATURES QUANTITATIVES : few differences

for col in df.select_dtypes('float'):
    plt.figure()
    sns.distplot(df_positive[col],label='positive')
    sns.distplot(df_negative[col],label='negative')
    plt.legend()
```
